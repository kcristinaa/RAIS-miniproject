```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
my_colors = ['#668B8B', '#CDAF95', '#CDB5CD', '#CAE1FF']
sns.set_style("whitegrid")
sns.set_palette(my_colors )
pd.options.display.max_colwidth=100
```

```python
age = pd.read_pickle('../data/feature_importances/age.pkl')
age = age.head(10)
gender = pd.read_pickle('../data/feature_importances/gender.pkl')
gender = gender.head(10)
```

```python
age = age.replace(["different_activity_types_4", "steps__fft_coefficient__attr_angle__coeff_24", "sleep_duration__fourier_entropy__bins_2", 
                   "steps__autocorrelation__lag_7", "sleep_duration__fft_coefficient__attr_angle__coeff_28", 
                   "sleep_duration__ar_coefficient__coeff_9__k_10",  "sleep_duration__fft_coefficient__attr_angle__coeff_22", 
                   "sleep_duration__count_below_mean", "stress_score__ratio_beyond_r_sigma__r_5", "steps__symmetry_looking__r_0.05_1.0"], ["activity_types", "steps__fft_24", 
                                           "sleep_fourier_entropy", "steps_autocorrelation", 
                                           "sleep__fft_28", "sleep__ar_coeff_9",
                                           "sleep__fft_22", "sleep_below_mean",
                                           "stress_ratio", " steps__symmetry"])
```

```python
fig = plt.gcf()
fig.set_size_inches(10, 10)
sns.set(font_scale = 2)
age.plot.barh(x='Feature', y='Value', title='Age Prediction', color='#668B8B')
plt.xlabel('Relative Importance')
plt.legend().set_visible(False)
plt.savefig("visualizations/age_feature_importance.png",bbox_inches='tight')
plt.savefig("visualizations/age_feature_importance.eps",bbox_inches='tight')
plt.savefig("visualizations/age_feature_importance.pdf",bbox_inches='tight')
```
