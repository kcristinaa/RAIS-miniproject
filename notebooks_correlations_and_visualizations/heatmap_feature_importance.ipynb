```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.options.display.max_colwidth=100
cmap='crest'
#cmap=['#fca55d', '#fee999', '#e3f399', '#9dd569', '#39a758']
```

```python
print(sns.color_palette("RdYlGn").as_hex())
```

```python
# demogaphics
age = pd.read_pickle('../data/feature_importances/age.pkl')
gender = pd.read_pickle('../data/feature_importances/gender.pkl')
bmi = pd.read_pickle('../data/feature_importances/bmi.pkl')
fitness = pd.read_pickle('../data/feature_importances/fitness.pkl')
# personality
agreeableness = pd.read_pickle('../data/feature_importances/agreeableness.pkl')
intellect = pd.read_pickle('../data/feature_importances/intellect.pkl')
extraversion = pd.read_pickle('../data/feature_importances/extraversion.pkl')
stability = pd.read_pickle('../data/feature_importances/stability.pkl')
conscientiousness = pd.read_pickle('../data/feature_importances/conscientiousness.pkl')
# mood
sema = pd.read_pickle('../data/feature_importances/sema.pkl')
stai = pd.read_pickle('../data/feature_importances/stai.pkl')
panas = pd.read_pickle('../data/feature_importances/panas.pkl')
```

```python
df=pd.concat([age, gender, bmi, fitness, agreeableness, intellect, conscientiousness, extraversion, stability, sema, stai, panas],axis=0).reset_index()
df = df.drop(columns='index')
# Rearranging the dataframe 
piv = pd.pivot_table(df, values=["Value"],index=["Label"], columns=["Feature"], fill_value=0)
sns.heatmap(piv, cmap=cmap)
plt.title('After 3rd preprocessing step - tsfresh')
plt.show()
```
